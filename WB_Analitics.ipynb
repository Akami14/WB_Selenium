```python
import pandas as pd
import numpy as np
import re
```

```python
path_1 = r'/content/cat_dump_19_11_2024_15_18.csv'
path_2 = r'/content/cat_dump_urls_19_11_2024_15_18.csv'
path_3 = r'/content/sub_cat_urls_19_11_2024_10_54.json'
```

```python
cat =  pd.read_csv(path_1, delimiter=';',  on_bad_lines='skip', header=None)
```

## Description
### 0 - url
### 1 - category name
### 2 - number of products in category
### 3 - 32 numbers of review

```python
cat = cat.iloc[:-34]
```

```python
cat = cat.dropna(axis=0)
```

```python
cat.describe()
```

```python
cat[2] = cat[2].astype(str)
```

```python
def clean_text(input_text):
    input_text = str(input_text)
    #Убираем специальные символы: избавляемся от всего, что не является "словами"
    clean = re.sub('[^а-яА-ЯёЁ0-9\s]', '', input_text) #
    try:
       clean = int(clean)
    except ValueError:
       clean = 0

    return clean
```

```python
cat[2] = cat[2].apply(clean_text)
```

```python
cat.head(6)
```

```python
cat['NOR'] = sum([cat[el] for el in np.arange(3,33)])/(33-3)
```

```python
cat['NOFPIC'] = cat[2].rename('NOFPIC')
```

```python
cat['NOR/NOFPIC'] = cat['NOR'] /cat['NOFPIC']
```

## 50 Potentially interesting category in descending arrange
I try to maximise NOR - number of reviews and  
 minimise NOFPIC - Number of product in category

```python
cat = cat[['NOR/NOFPIC', 1, 0]].sort_values(by=['NOR/NOFPIC'], ascending=False)
cat = cat.loc[cat['NOR/NOFPIC'] != np.inf]
cat.head(50)
```
